Basado en contenido

```python
#Importe de librerias
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Información de géneros de movies
movies_df = pd.read_csv('/content/drive/MyDrive/movies.csv')
#Historia de usuarios
ratings_df = pd.read_csv('/content/drive/MyDrive/ratings.csv')

movies_df.head()

```

**Acondicionamos Géneros de Movie**

```python
#Indexar una columna, la llamaremos year, extraer la fecha de la columna title
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
movies_df.head()
```

```python
#Removemos el parentesis del año
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
movies_df.head()
```

```python
#Remover del la columna title el año con la función replace un espacio en blanco en el lugar del año
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')
movies_df.head()
```

```python
#Aplicar la función strip para eliminar el espacio en blanco que aparezca al final del registro de title
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

```python
#Modificar los registros genres que tenga separados por |, en una lista
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

**Copiado y generación de géneros de Movie**

```python
##Copiado de movie para generar un nuevo DATAFRAMES, en donde se alojará la codificación con el enfoque One-Hot
genero_peli_df = movies_df.copy()

#Hacer el recorrido y asignar 1 a si la película pertecene a determinado género
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        genero_peli_df.at[index, genre] = 1

#Llenar con 0 los espacios en blanco, es decir si la película no pertenence a determinado género
genero_peli_df = genero_peli_df.fillna(0)
genero_peli_df.head()
```

**Acondicionamos Historia de Usuarios**

```python
ratings_df.head()
```

```python
#Borramos la columna timestamp
ratings_df = ratings_df.drop('timestamp', 1)
ratings_df.head()
```

**<h2>Sistema de recomendación basado en contenido**

***Perfil de Usuario***

Creamos un vector de entrada de usuarios con las puntuaciones de cada pelicula

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5},
            {'title': 'Avatar', 'rating': 5}
         ] 
#Generamos un DataFrames a partir de la lista
inputMovies = pd.DataFrame(userInput)
inputMovies
```

**Agreguemos movieId a la matriz**

```python
#Filtramos las salidas de peliculas en la columna title
#isin comprueba los valores de entrada
#tolist convierte en lista
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]

#Fusionamos inpuId-inputMovies
inputMovies = pd.merge(inputId, inputMovies)

#Descartar información que no usaremos del marco de datos de entrada
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)

inputMovies
```

```python
#Filtrado de salidas de usuario para movie
userMovies = genero_peli_df [genero_peli_df ['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

**Matriz Movie**

```python
#Restablecer el índice para evitar problemas futuros
userMovies = userMovies.reset_index(drop=True)
#Eliminación las columnas movieId, title, genres y year para evitar conflictos futuros
userGenreTable = userMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
userGenreTable
```

```python
inputMovies['rating']
```

**Pesos de cada género**

```python
#Conseguir los pesos para la matriz ponderada por cada género
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
#The user profile
userProfile
```

```python
#Obtenemos los géneros de cada pelicula el conjunto de datos
genreTable = genero_peli_df.set_index(genero_peli_df['movieId'])
#Borramos columnas incesarias para esta visualización
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)

#Muestro la tabla
genreTable.head()
```

```python
genreTable.shape
```

```python
#Average de los pesos de los géneros

#genreTable= tabla generada con todos los codigo de los géneros por cada película
#userProfile= pesos por géneros.
#Se multiplican las dos matrices y luego se suma todas la columnas, es decir que cada 
#pelicula tendrá un valor que será plasmado en la matriz de recomendación.

recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head(10)
```

```python
#Ordenamos de manera descendente para saber cuales puntajes son mayores
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
#Visualizar los valores
recommendationTable_df.head(10)
```

**Tabla de Recomendación final**

```python
#Se muestra la matriz de recomendaciones
#Las primeras 5 películas más recomendadas son:
movies_df.loc[movies_df['movieId'].isin(recommendationTable_df.head().keys())]
```
